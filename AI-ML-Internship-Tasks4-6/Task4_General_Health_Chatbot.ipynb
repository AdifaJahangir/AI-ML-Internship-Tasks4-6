```python
from transformers import pipeline

chatbot = pipeline(
    "text-generation",
    model="microsoft/DialoGPT-medium"
)
```

```python
dangerous_words = [
    "suicide",
    "kill myself",
    "overdose",
    "self harm"
]

def safety_check(text):

    text = text.lower()

    for word in dangerous_words:
        if word in text:
            return False

    return True


def health_chatbot(question):

    question = question.lower()

    if "sore throat" in question:
        return "A sore throat is commonly caused by viral infections, allergies, or irritation. If symptoms persist, consult a healthcare professional."

    elif "blood pressure" in question:
        return "Regular exercise, reducing salt intake, maintaining a healthy weight, and following medical advice can help manage blood pressure."

    elif "headache" in question:
        return "Headaches may occur due to stress, dehydration, lack of sleep, or illness."

    elif "water" in question:
        return "Most adults should drink sufficient water daily, but needs vary based on age, activity level, and climate."

    elif "fever" in question:
        return "Fever is often a sign that the body is fighting an infection. Seek medical advice if the fever is high or persistent."

    else:
        return "I can provide general health information. Please consult a healthcare professional for specific medical advice."
```

Chat Loop

```python
while True:

    user = input("Ask a health question: ")

    if user.lower() == "exit":
        print("Chatbot Closed.")
        break

    if not safety_check(user):
        print("Please contact a healthcare professional immediately.")
        continue

    print("\nChatbot:", health_chatbot(user))
    print()
```
